```python
import networkx as nx
import pandas as pd
```

```python
df = pd.read_csv("musae_facebook_edges.csv")
print(df)
```

```python
list1 = df['id_1'].tolist()
list2 = df['id_2'].tolist()

graph = nx.Graph()
for i in range(0, len(list1)):
  graph.add_edge(list1[i], list2[i])
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(20, 10))
nx.draw_networkx(graph, node_size=1000, width=1, node_color='yellow')
#plt.savefig('books_read.png')
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(40, 40))
nx.draw_networkx(graph, node_size=100, width=1, node_color='yellow')
#plt.savefig('books_read.png')
```
